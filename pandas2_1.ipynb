Ex2 - Filtering and Sorting Data
This time we are going to pull data directly from the internet **bold text**

```python
# Step 1. Import the necessary libraries

import pandas as pd
```

```python
# Step 2. Import the dataset from this address.

df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv")
df.head()
```

```python
# Step 3. Assign it to a variable called euro12.

euro12 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv")
euro12.head()
```

```python
# Step 4. Select only the Goal column.

# Select the 'Goal' column using .loc[]
goal_column = euro12.loc[:, 'Goals']

# Display the 'Goal' column
print(goal_column)
```

```python
# Step 5. How many team participated in the Euro2012?

# Count the number of unique teams in the 'Team' column
num_teams = euro12['Team'].nunique()

print("Number of teams participated in Euro2012:", num_teams)
```

```python
# Step 6. What is the number of columns in the dataset?

# Use shape to find the number of columns in the dataset
# Get the shape of the DataFrame
num_rows, num_columns = euro12.shape

print("Number of columns in the dataset:", num_columns)
```

```python
# Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

# Select the specified columns and assign them to a new DataFrame called discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]

# Display the first few rows of the discipline DataFrame
discipline.head()
```

```python
# Step 8. Sort the teams by Red Cards, then to Yellow Cards

# Sort the discipline DataFrame by 'Red Cards' and then by 'Yellow Cards'
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'])

# Display the sorted DataFrame
print(discipline_sorted)
```

```python
# Step 9. Calculate the mean Yellow Cards given per Team

# To calculate the mean Yellow Cards given per team
mean_yellow_cards_per_team = discipline['Yellow Cards'].mean()

print("Mean Yellow Cards given per Team:", mean_yellow_cards_per_team)
```

```python
# Step 10. Filter teams that scored more than 6 goals

# To filter teams that scored more than 6 goals
teams_more_than_6_goals = euro12[euro12['Goals'] > 6]

# Display the filtered DataFrame
print(teams_more_than_6_goals)
```

```python
# Step 11. Select the teams that start with G

# Filter teams that start with the letter 'G'
teams_starting_with_G = euro12[euro12['Team'].str.startswith('G')]

# Display the filtered DataFrame
print(teams_starting_with_G)
```

```python
# Step 12. Select the first 7 columns

# Select the first 7 columns
first_7_columns = euro12.iloc[:, :7]

# Display the selected columns
print(first_7_columns)
```

```python
# Step 13. Select all columns except the last 3.

# Select all columns except the last 3
all_except_last_3_columns = euro12.iloc[:, :-3]

# Display the selected columns
print(all_except_last_3_columns)
```

```python
# Step 14. Present only the Shooting Accuracy from England, Italy and Russia
# Filter rows for England, Italy, and Russia
selected_teams = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])]

# Select the 'Team' and 'Shooting Accuracy' columns
shooting_accuracy_selected_teams = selected_teams[['Team', 'Shooting Accuracy']]

# Display the DataFrame
print(shooting_accuracy_selected_teams)
```
